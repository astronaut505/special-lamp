```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

ticker = "^GSPC"
start_date = "1970-01-01"
end_date = "2023-06-11"

data = yf.download(ticker, start=start_date, end=end_date)
data["E_P_ratio"] = 1 / data["Close"]
data["E_P_ratio_lagged"] = data["E_P_ratio"].shift(1)
data["Returns_lagged"] = data["Close"].pct_change().shift(1)
data = data.dropna()
```

```python
X = sm.add_constant(data["E_P_ratio_lagged"])
y = data["Returns_lagged"]

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```

```python
data["E_P_ratio_rolling_median"] = data["E_P_ratio"].rolling(window=120, min_periods=1).median()

plt.figure(figsize=(12, 6))
plt.plot(data.index, data["E_P_ratio"], label="E/P ratio")
plt.plot(data.index, data["E_P_ratio_rolling_median"], label="Rolling median")
plt.xlabel("Date")
plt.ylabel("E/P ratio")
plt.title("Chart 1: E/P ratio and Rolling Median")
plt.legend()
plt.show()
```

```python
percentile_95 = data["E_P_ratio"].rolling(window=120, min_periods=1).quantile(0.95)
percentile_5 = data["E_P_ratio"].rolling(window=120, min_periods=1).quantile(0.05)
data["E_P_ratio_percentile_range"] = percentile_95 - percentile_5

plt.figure(figsize=(12, 6))
plt.plot(data.index, data["E_P_ratio"], label="E/P ratio")
plt.fill_between(data.index, percentile_5, percentile_95, alpha=0.2, label="Percentile Range")
plt.xlabel("Date")
plt.ylabel("E/P ratio")
plt.title("Chart 2: E/P ratio and Percentile Range")
plt.legend()
plt.show()
```

```python
data["Returns_10yr"] = data["Adj Close"].pct_change(periods=10)
data["PE_Ratio"] = data["Adj Close"] / data["Close"]

data_sorted_by_pe = data.sort_values("PE_Ratio")

plt.figure(figsize=(12, 6))
plt.scatter(data_sorted_by_pe["PE_Ratio"], data_sorted_by_pe["Returns_10yr"])
plt.xlabel("Starting P/E Ratio")
plt.ylabel("10-year Returns")
plt.title("Chart 1: U.S. Equity 10-year Returns Sorted by Starting P/E Ratio")
plt.show()

data["Returns_60yr"] = data["Adj Close"].pct_change(periods=60)

data_sorted_by_valuation = data.sort_values("Returns_60yr")

plt.figure(figsize=(12, 6))
plt.scatter(data_sorted_by_valuation["Returns_60yr"], data_sorted_by_valuation["Adj Close"])
plt.xlabel("Starting Valuation based on Rolling 60-year Window")
plt.ylabel("U.S. Equity Returns")
plt.title("Chart 2: U.S. Equity Returns Sorted by Starting Valuation based on Rolling 60-year Window")
plt.show()
```

```python
data["E_P_ratio_percentile_range"] = percentile_95 - percentile_5

# Calculate lagged returns
data["Returns_lagged"] = data["Close"].pct_change().shift(1)

# Drop rows with missing values again
data = data.dropna()

# Calculate the market timing strategy's equity weight
data["Equity_Weight"] = 100 + (data["E_P_ratio"] - data["E_P_ratio_rolling_median"]) / data["E_P_ratio_percentile_range"]

# Limit the equity weight between 50% and 150%
data["Equity_Weight"] = np.clip(data["Equity_Weight"], 50, 150)

# Calculate the cash weight as the complement of the equity weight
data["Cash_Weight"] = 100 - data["Equity_Weight"]

# Calculate the market timing strategy's returns
data["Strategy_Returns"] = data["Returns_lagged"] * (data["Equity_Weight"] / 100) + data["Returns_lagged"].shift(1) * (data["Cash_Weight"] / 100)

# Calculate the cumulative returns of the market timing strategy
data["Cumulative_Strategy_Returns"] = (1 + data["Strategy_Returns"]).cumprod()

# Calculate the cumulative returns of the underlying broad market index (buy-and-hold strategy)
data["Cumulative_BuyHold_Returns"] = (1 + data["Returns_lagged"]).cumprod()

# Plot the cumulative returns of the market timing strategy and the buy-and-hold strategy
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Cumulative_Strategy_Returns"], label="Market Timing Strategy")
plt.plot(data.index, data["Cumulative_BuyHold_Returns"], label="Buy-and-Hold Strategy")
plt.xlabel("Date")
plt.ylabel("Cumulative Returns")
plt.title("Comparison of Market Timing Strategy and Buy-and-Hold Strategy")
plt.legend()
plt.show()
```

```python
# Calculate the annualized returns and standard deviation of the market timing strategy
strategy_annual_returns = data["Strategy_Returns"].mean() * 252
strategy_annual_volatility = data["Strategy_Returns"].std() * np.sqrt(252)

# Calculate the annualized returns and standard deviation of the buy-and-hold strategy
buyhold_annual_returns = data["Returns_lagged"].mean() * 252
buyhold_annual_volatility = data["Returns_lagged"].std() * np.sqrt(252)

# Calculate the risk-free rate (e.g., 10-year Treasury yield)
risk_free_rate = 0.03

# Calculate the Sharpe ratio for the market timing strategy and the buy-and-hold strategy
strategy_sharpe_ratio = (strategy_annual_returns - risk_free_rate) / strategy_annual_volatility
buyhold_sharpe_ratio = (buyhold_annual_returns - risk_free_rate) / buyhold_annual_volatility

# Print the performance metrics
print("Market Timing Strategy:")
print("Annualized Returns:", strategy_annual_returns)
print("Annualized Volatility:", strategy_annual_volatility)
print("Sharpe Ratio:", strategy_sharpe_ratio)

print("\nBuy-and-Hold Strategy:")
print("Annualized Returns:", buyhold_annual_returns)
print("Annualized Volatility:", buyhold_annual_volatility)
print("Sharpe Ratio:", buyhold_sharpe_ratio)
```

```python
from scipy.stats import ttest_ind

# Perform a t-test to compare the mean returns of the market timing strategy and the buy-and-hold strategy
t_stat, p_value = ttest_ind(data["Strategy_Returns"], data["Returns_lagged"], alternative='greater')

# Check if the market timing strategy outperformed the buy-and-hold strategy at a significant level (e.g., p-value < 0.05)
if p_value < 0.05:
    print("The market timing strategy outperformed the buy-and-hold strategy.")
else:
    print("The market timing strategy did not outperform the buy-and-hold strategy.")
```
